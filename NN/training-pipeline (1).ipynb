```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/kaggle/input/data-for-breast-cancer/data.csv")
print(df.head())
```

```python
df.columns
```

```python
df.drop(columns = ['Unnamed: 32', 'id'], inplace = True)
```

```python
print(df.columns)
```

```python
print(df.head())
```

```python
x_train, x_test, y_train , y_test= train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size = 0.2)
```

```python
print(y_train.name)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

```python
y_train
```

```python
y_test
```

```python
import torch

# Convert NumPy arrays to PyTorch tensors
x_train_tensor = torch.from_numpy(x_train).to(torch.float32)
x_test_tensor = torch.from_numpy(x_test).to(torch.float32)
y_train_tensor = torch.from_numpy(y_train).to(torch.float32)
y_test_tensor = torch.from_numpy(y_test).to(torch.float32)

# Verify dtype
print(x_train_tensor.dtype)  # Should print torch.float32
print(y_train_tensor.dtype)  # Should print torch.float32
```

```python
import torch
import torch.nn as nn

class mySimpleNN(nn.Module):
    def __init__(self, number_features):
        super().__init__()
        self.linear = nn.Linear(number_features, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, features):
        out = self.linear(features)
        y_pred = self.sigmoid(out)
        return y_pred
```

```python
learning_rate = 0.1
epochs = 25
```

```python
loss_function = nn.BCELoss()
```

```python
#create model
model = mySimpleNN(x_train_tensor.shape[1])
print(x_train_tensor.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
for epoch in range(epochs):
    #forword pass
    y_pred = model(x_train_tensor)
    loss = loss_function(y_pred,y_train_tensor.view(-1,1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()
    print(f'epoch : {epoch+1}, loss:{loss.item()}')
```

```python
print(model.linear.weight)
```

```python
with torch.no_grad():
    y_pred = model.forward(x_train_tensor)
    y_pred = (y_pred>0.5).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(accuracy)
```
